```
import numpy as np
from math import sqrt
from collections import Counter
import matplotlib.pyplot as plt
from matplotlib import style
import warnings
from math import ceil


k=int(input("k: "))
n=int(input("number of datapoints: "))
l=input("l1 or l2: ")

def k_nearest_neighbors(data, predict, k, l):
    if len(data) >= k:
        warnings.warn('K is recommended to be greater than number of classes')

    dists = []
    if l == "l2":
        for g in data:
            for f in data[g]:
                l2_dist = abs((f[0] * f[0]) - (predict[0] * predict[0])) + abs(
                    (f[1] * f[1]) - (predict[1] * predict[1]))
                dists.append([l2_dist, g])
    else:
        for g in data:
            for f in data[g]:
                l1_dist = np.linalg.norm(np.array(f) - np.array(predict))
                dists.append([l1_dist, g])

    c = [i[1] for i in sorted(dists)[:k]]
    vote_result = Counter(c).most_common(1)[0][0]
    return vote_result


dataset = {'b': [], 'g': [], "c": []}
dataset_copy = {'k': [], 'r': [], "m": []}

from random import randint
ar1=["b","g","c"]
ar2=["k","r","m"]
for i in range(n):
    p=randint(0,2)
    x=randint(0,100)/10
    y=randint(0,100)/10
    dataset[ar1[p]].append([x,y])
    dataset_copy[ar2[p]].append([x,y])
    
xmax = 0
ymax = 0
for i in dataset.keys():
    for j in dataset[i]:
        x, y = j[0], j[1]
        xmax = max(x,ceil(xmax))
        ymax = max(y,ceil(ymax))



for i in range(0, (xmax + 1) * 10):
    for j in range(0, (ymax + 1) * 10):
        x = i / 10
        y = j / 10
        new_f = [x, y]
        result = k_nearest_neighbors(dataset, new_f, k, l)
        plt.scatter(new_f[0], new_f[1], s=100, color=result)

[[plt.scatter(ii[0], ii[1], s=100, color=i) for ii in dataset_copy[i]] for i in dataset_copy]

plt.title('K-NN classifier')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
